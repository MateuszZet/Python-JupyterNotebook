```python
import math

def multiply(number,number2):
    if (str(number).lower() == "bunny" or str(number).lower() == "rabbit") and (str(number2).lower() == "bunny" or str(number2).lower() == "rabbit") :
        print("\nI am just a dumb bunny, but we are good at multiplying.")
        a = {"Parent1" : number.title(), "Parent2" : number2.title(), "Kid1" : "Bunny" , "Kid2" : "Bunny" , "Kid3" : "Bunny" , "Kid4" : "Bunny"  }
    elif isinstance(number, str) and isinstance(number2, str):
        a = {"Product" : number, "Info" : number2 }
    elif isinstance(number, str) and (isinstance(number2, int) or isinstance(number2, float)) : 
        a = {"Product" : number, "Quantity" : math.floor(number2)}
    elif isinstance(number2, str) and (isinstance(number, int) or isinstance(number, float)) :
        a = {"Product" : number2, "Quantity" : math.floor(number)}
    else:
        a1 = float(number) * float(number2)
        a2 = math.floor(a1)
        a = {"Rounded" : a2 , "Precise" : a1 }
        
    return a


print(multiply(5, 43))
print(multiply(43.76, 22))
print(multiply(4.44, 5.43))
print(multiply("Paper", "white"))
print(multiply("Pencil", 2.2))
print(multiply("T-Shirt", 123))
print(multiply("bUNNY", "bunny"))
print(multiply("Rabbit", "raBbit"))
print(multiply("Rabbit", "Bunny"))
print(multiply("bUnnY", "RABBIT"))
```

```python
#fibonacci generator
def fib(num):
    a, b = 0,1
    for i in range(0, num):
        yield "{}: {}".format(i+1,a)
        a, b = b, a + b
        
for item in fib(33):
    print(item)
```

```python
#fibonacci function
def fib_fun(num):
    a = 0
    b = 1

    for i in range(0,num):
        print(str(i+1) + ": " + str(a))
        a = b
        b = a + b
      
fib_fun(154)
```

```python
#function
def add(product):
    if isinstance(product, str):
        d = product.upper()
    else:
        d = str(product)
    #print("Adding " + d)
    return d


product_list = ['milk', 'butter', 'tea']
cp_product_list = product_list.copy()

print("This is copy of original list: " + str(cp_product_list))
print("This is original list: " + str(product_list))
print("\n")

#Changing all strings in original list to upper
for idx, item in enumerate(cp_product_list):
    print(item + " going UPPER!")
    cp_product_list[idx] = item.upper()

print("\n")
print("This is copy of original list: " + str(cp_product_list))
print("This is original list: " + str(product_list))
print("\n")
print("Adding int and string to original list using add function.")
print("\n")

cp_product_list.append(add(123))
cp_product_list.append(add("pIzZa"))


print("This is copy of original list: " + str(cp_product_list))
print("This is original: " + str(product_list))
```

```python
#generator
def kwackgenerator(x):
    d = "kwack "
    i = 1
    while i < x:
        yield d*i
        i += 1


x = kwackgenerator(5)

print("First 2 items: \n")
f = next(x)
print(f)
f = next(x)
print(f)

print("\nSTART of FOR loop: ")

for item in x:
    print(item)
```
